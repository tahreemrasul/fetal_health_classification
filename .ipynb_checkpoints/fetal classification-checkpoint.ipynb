```python
## independent and dependent features
X = df.iloc[:,:-1]
y = df.iloc[ :, -1:]
```

```python
X.head()
```

```python
y.head()
```

```python
## feature importance
from sklearn.ensemble import ExtraTreesClassifier
model=ExtraTreesClassifier()
model.fit(X,np.ravel(y))
```

```python
print(model.feature_importances_)
```

```python
## plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
top_features=feat_importances.nlargest(7)
top_features.plot(kind='barh')
plt.show()
```

```python
# pick the most important features for the training dataset
X_top=X[top_features.index]
X_top.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X_top, y, test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier()
```

```python
## Hyperparameter tuning

# Number of trees in random forest
n_estimators=[int(x) for x in np.linspace(start=100, stop=1200, num=12)]
# Number of features to consider at every split
max_features=['auto', 'sqrt']
# Maximum number of levels in tree/depth of tree
max_depth=[int(x) for x in np.linspace(start=5, stop=30, num=6)]
# Minimum number of samples required to split a node
min_samples_split=[2,5,10,15,100]
# Minimum number of samples required at each leaf node
min_samples_leaf=[1,2,5,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
print(random_grid)
```

```python
# find the best hyperparameters using the random grid
# base model creation
rf = RandomForestClassifier()
```

```python
rf_random = RandomizedSearchCV(estimator=rf, param_distributions=random_grid, scoring='accuracy', n_iter=10, cv=5, verbose=2, random_state=42, n_jobs=1)
```

```python
top_estimators = rf_random.fit(X_train, np.ravel(y_train))
```

```python
print(top_estimators.best_params_)
```

```python
print(top_estimators.best_estimator_)
```

```python
predictions=rf_random.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy: ", accuracy_score(np.ravel(y_test), predictions)*100, "%")
```

```python
# confusion matrices can be used to see how well the model predicts each outcome
from sklearn.metrics import confusion_matrix
conmat=confusion_matrix(np.ravel(y_test), predictions)
val=np.mat(conmat)
fetal_health_classes=list(np.unique(y_train))
df_cm = pd.DataFrame(val, index=fetal_health_classes, columns=fetal_health_classes)
print(df_cm)
```

```python
# create a heatmap to plot the confusion matrix
plt.figure(figsize=(10,10))
heat_map_predictions=sns.heatmap(df_cm, annot=True, cmap='Blues', fmt='g')
heat_map_predictions.yaxis.set_ticklabels(heat_map_predictions.yaxis.get_ticklabels(), rotation=0, ha='right')
heat_map_predictions.xaxis.set_ticklabels(heat_map_predictions.xaxis.get_ticklabels(), rotation=45, ha='right')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Fetal Health Random Forest Classifier Results')
plt.show()
```

```python
# display the results as percentages of the total number
df_cm = df_cm.astype('float')/df_cm.sum(axis=1)[:, np.newaxis]
plt.figure(figsize=(10,10))
heat_map_predictions=sns.heatmap(df_cm, annot=True, cmap='Blues', fmt='g')
heat_map_predictions.yaxis.set_ticklabels(heat_map_predictions.yaxis.get_ticklabels(), rotation=0, ha='right')
heat_map_predictions.xaxis.set_ticklabels(heat_map_predictions.xaxis.get_ticklabels(), rotation=45, ha='right')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Fetal Health Random Forest Classifier Results (%)')
plt.show()
```
